```python
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import MinMaxScaler
from keras.models import Sequential
from keras.layers import LSTM, GRU, SimpleRNN, Dense, Dropout, Bidirectional
from keras.callbacks import EarlyStopping, ReduceLROnPlateau
from sklearn.metrics import mean_squared_error, mean_absolute_error
from math import sqrt

# Step 1: Load and Preprocess Data
def load_data_aal(file_path, time_steps=60):
    data = pd.read_csv(file_path)
    data['Close'] = data['Close'].astype(float)

    # Plot histogram for data distribution
    plt.figure(figsize=(12, 6))
    sns.histplot(data['Close'], kde=True, color='blue', bins=50)
    plt.title("Distribution of Closing Prices")
    plt.xlabel("Closing Price")
    plt.ylabel("Frequency")
    plt.grid()
    plt.show()

    # Scale the 'Close' column
    scaler = MinMaxScaler(feature_range=(0, 1))
    scaled_data = scaler.fit_transform(data[['Close']])

    # Ensure there is enough data to create sequences
    if len(scaled_data) <= time_steps:
        raise ValueError(f"Dataset is too small to create sequences with {time_steps} time steps. Dataset size: {len(scaled_data)}")

    # Create sequences for training/testing
    X, y = [], []
    for i in range(time_steps, len(scaled_data)):
        X.append(scaled_data[i-time_steps:i, 0])
        y.append(scaled_data[i, 0])

    X, y = np.array(X), np.array(y)
    X = X.reshape(X.shape[0], X.shape[1], 1)  # Reshape for RNN input
    return X, y, scaler, data

# Step 2: Define RNN Model
def build_rnn_model(model_type, input_shape, units=50):
    model = Sequential()
    if model_type == "LSTM":
        model.add(Bidirectional(LSTM(units=units, return_sequences=True, recurrent_dropout=0.2), input_shape=input_shape))
        model.add(Bidirectional(LSTM(units=units//2, recurrent_dropout=0.2)))
    elif model_type == "GRU":
        model.add(Bidirectional(GRU(units=units, return_sequences=True, recurrent_dropout=0.2), input_shape=input_shape))
        model.add(Bidirectional(GRU(units=units//2, recurrent_dropout=0.2)))
    elif model_type == "SimpleRNN":
        model.add(SimpleRNN(units=units, activation='relu', return_sequences=False, input_shape=input_shape))
    model.add(Dropout(0.3))
    model.add(Dense(1))
    model.compile(optimizer="adam", loss="mean_squared_error")
    return model

# Step 3: Evaluate Models with Validation Split
def evaluate_models_with_validation(X_train, y_train, X_val, y_val, X_test, y_test, models=["SimpleRNN", "LSTM", "GRU"]):
    results = {}
    for model_type in models:
        print(f"Training {model_type}...")
        model = build_rnn_model(model_type, X_train.shape[1:], units=100)  # Increased units for complexity

        # Apply early stopping and learning rate reduction
        early_stopping = EarlyStopping(monitor='val_loss', patience=10, restore_best_weights=True)
        reduce_lr = ReduceLROnPlateau(monitor='val_loss', factor=0.5, patience=5, min_lr=1e-6)

        # Train the model using the training and validation sets
        history = model.fit(
            X_train, y_train,
            epochs=20,
            batch_size=32,
            validation_data=(X_val, y_val),  # Validation data
            callbacks=[early_stopping, reduce_lr],
            verbose=1
        )

        # Predict and calculate performance metrics
        train_predictions = model.predict(X_train)
        val_predictions = model.predict(X_val)
        test_predictions = model.predict(X_test)

        # Plot predictions
        plt.figure(figsize=(14, 7))
        plt.plot(range(len(y_train)), y_train, label='Actual Training Data', color='blue')
        plt.plot(range(len(y_train)), train_predictions.flatten(), label='Training Predictions', color='orange')
        plt.plot(range(len(y_train), len(y_train) + len(y_val)), y_val, label='Actual Validation Data', color='green')
        plt.plot(range(len(y_train), len(y_train) + len(y_val)), val_predictions.flatten(), label='Validation Predictions', color='red')
        plt.plot(range(len(y_train) + len(y_val), len(y_train) + len(y_val) + len(y_test)), y_test, label='Actual Test Data', color='purple')
        plt.plot(range(len(y_train) + len(y_val), len(y_train) + len(y_val) + len(y_test)), test_predictions.flatten(), label='Test Predictions', color='brown')
        plt.legend()
        plt.title(f"{model_type} Model Predictions")
        plt.xlabel("Time Steps")
        plt.ylabel("Scaled Values")
        plt.show()

        rmse = sqrt(mean_squared_error(y_test, test_predictions))
        mae = mean_absolute_error(y_test, test_predictions)
        mape = np.mean(np.abs((y_test - test_predictions.flatten()) / y_test)) * 100
        accuracy = 100 - mape

        results[model_type] = {"RMSE": rmse, "MAE": mae, "MAPE (%)": mape, "Accuracy (%)": accuracy}

    return results

# Step 4: Main Execution
file_path = r"AAL.csv"  # Update to your file path
time_steps = 60

# Load data
X, y, scaler, data = load_data_aal(file_path, time_steps)

# Split into training, validation, and testing sets
train_size = int(len(X) * 0.8)
val_size = int(train_size * 0.2)

X_train, y_train = X[:train_size], y[:train_size]
X_test, y_test = X[train_size:], y[train_size:]

X_train_actual, X_val = X_train[:-val_size], X_train[-val_size:]
y_train_actual, y_val = y_train[:-val_size], y_train[-val_size:]

# Evaluate models
results = evaluate_models_with_validation(X_train_actual, y_train_actual, X_val, y_val, X_test, y_test)

# Print Results
print("Model Performance Comparison with Validation:")
performance_df = pd.DataFrame(results).T[["RMSE", "MAE", "MAPE (%)", "Accuracy (%)"]]
print(performance_df)
```
